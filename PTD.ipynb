# 2. Importing the Required Libs

```python
import PyPDF2
import pdfplumber
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
filepath = 'stock_market_data.pdf' # Defining the path of the PDF
```

# 3.Reading the PDF DOC WITH THE PyPDF2 lib

```python
with open(filepath, 'rb') as file:
    reader = PyPDF2.PdfReader(file)
    print(f"Number of pages : {len(reader.pages)}")
    page = reader.pages[0]
    text = page.extract_text()
    print("Text from PYPDF: \n", text)
```

## 4. Reading the PDF with PDFplumber with Table

```python
with pdfplumber.open(filepath) as pdf:
    first_page = pdf.pages[0]
    text = first_page.extract_text()
    print("Raw Text : \n", text)
```

```python
table = first_page.extract_table() #Extracting the Table
print("Table Data: \n", table)
```

# 5. Converting the DataFrame

```python
if table:
    df = pd.DataFrame(table[1:], columns = table[0])
    print(df)
```

```python
# Ensure 'Today' column is treated as a string before applying string operations
df['Today'] = df['Today'].astype(str).str.replace('%', '').str.strip().astype(float)

# Clean the 'Sector' column
df['Sector'] = df['Sector'].str.strip()

# Set up the figure
plt.figure(figsize=(16, 9))
sns.set_theme(style='whitegrid')

# Create bar plot for each sector's growth
sns.barplot(data=df, x='Sector', y='Today', palette='Set2')

# Labels and aesthetics
plt.title("Sector-wise Stock Growth Today", fontsize=16)
plt.xlabel("Sector", fontsize=12)
plt.ylabel("Growth (%) Today", fontsize=12)
plt.xticks(rotation=45)
plt.axhline(0, color='gray', linestyle='--')

# Display the plot
plt.tight_layout()
plt.show()
```
